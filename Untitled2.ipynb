```python
# from storage to access all the file
from glob import glob
train_path="/content/Training"
validation_path="/content/validation"
# initialise  glob to interact with storage
image_files = glob(train_path +'/*/*/.jp*g')
validateing_files = glob(validation_path +'/*/*/.jp*g')
```

```python
# creating imagedatagenerator object,to access preprocessor function
# vgg16
generator = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input,
                                                           rotation_range=20,
                                                           shear_range=0.1,
                                                           zoom_range=0.2,
                                                           horizontal_flip=True,
                                                           vertical_flip=True)
train_generator= generator.flow_from_directory(train_path,
                              target_size=(224,224),
                              class_mode='categorical',
                              batch_size=64)
validation_generator= generator.flow_from_directory(validation_path,
                              target_size=(224,224),
                              class_mode='categorical',
                              batch_size=64)
```

```python
# load the model

#include_top= false
#include_top = True , only prediction

vgg= tf.keras.applications.vgg16.VGG16(input_shape=(224,224,3),
                                  weights='imagenet',
                                  include_top = False)
for layer in vgg.layers:
  layer.trainable =False


# flattening layer

flatten = tf.keras.layers.Flatten()

#output layer

output_layer = tf.keras.layers.Dense(7,activation = "softmax")






```

```python
vgg.summary()
```

```python
model=tf.keras.models.Sequential([vgg,
                           flatten,
                            output_layer])
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
```

```python
model.fit(train_generator,
          validation_data=validation_generator,
          epochs=10,
          batch_size=32)
```

```python
model.save("fruits360.h5")
```
